```python
# writing unique values
def remove_duplicates(string):
    # Convert the string to lowercase and split it into a list of words
    words = string.lower().split()
    
    # Create an empty list to store unique words
    unique_words = []
    
    # Iterate over the list of words
    for word in words:
        # If the word is not already in the list of unique words, add it
        if word not in unique_words:
            unique_words.append(word)
            
    return unique_words
```

## ** Kwargs:
When We don't know, how many keyword arguments are going to pass in te a function is called **kwargs.
**kwargs allows us to pass any number of keyword arguments.

Keyword arguments mean that they contain a key-value pair, like a Python dictionary.

```python
def f6(**kwargs):
    return kwargs
```

```python
f6(b = 4,c = 5,d =  6,e = 6 , n = [3,4,45,5,6,6,6],f="ravi")
```
