# - Types constructors

- 1.Parameterized Constructor
A parameterized constructor takes one or more arguments in addition to  self . It 
allows for the initialization of an object with specific values provided at the time of 
object creation


- 2.Non-Parameterized Constructor

 Initializes attributes with default values. It is 
useful when you want to provide default settings or when initialization does not

````python

class netflix:

    new_show = 22

    def __init__(self,mail, password,service,amount ): # 1.Parameterized Constructor
        self.mail = mail
        self.password = password
        self.service = service
        self.amount =  amount 

        


    def get_password(self):
        return self.service, self.amount
````

- in the above example we are defined that parameters  within ```__inti__``` and  the same values are defined below as instance variables is called paramererized construtor

# Instance Methods

Instance methods are the most common methods in a class. They operate on 
instance data and can access and modify instance attributes. They are defined 
with at least one parameter, typically named  ```self``` , which refers to the instance on

````python
class car() :
    def __init__(self , milage, year , make , model):
        self.milage = milage
        self.year = year
        self.make = make
        self.model = model

    def car_m(self):
      print("milage:",self.milage)

      print("year:",self.year)

    def car_details(self):
      print("milage:",self.milage)
      print("year;",self.year)
      print("model:",self.model)
````

# Class Methods

Class methods operate on the class itself, rather than on instances of the class. 
They are defined using the  @classmethod  decorator and take a  ```cls```  parameter, which 
refers to the class, not the instance. They are used to access or modify class-level 
data, such as class attributes

````python
class netflix:

    new_show = "Pushpa"

    def __init__(self,mail, password):
        self.mail = mail
        self.password = password
        self.service = "freee"
        self.amount = 0 

        


    def get_password(self):
        return self.password
    

    @classmethod
    def get_new_show(cls):
        return cls.new_show
    
    @classmethod
    def set_new_show (cls,show_name):
        cls.new_show = show_name
    
````

````python
netflix.get_new_show()  # accessing class method using class without object 
````

````python
myuser=netflix("laxman",23)     # creating object of class netflix 
myuser.get_new_show()  
````

````python
netflix.set_new_show("part6")
print(netflix.get_new_show())   
````

````python
myuser=netflix("pirn",23)
myuser.set_new_show(45)
myuser.get_new_show()
````
