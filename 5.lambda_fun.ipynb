## What are list comprehensions?
List comprehensions are a tool for transforming one list (any iterable actually) into another list.

During this transformation, elements can be conditionally included in the new list and each element can be transformed as needed.

List comprehensions in Python are great, but mastering them can be tricky because they don’t solve a new problem: they just provide a new syntax to solve an existing problem.

The iterator part iterates through each member number of the input sequence list_of_numbers.

The predicate checks if the member is even.

## the generator doesn’t produce all the values at the same time, it saves memory if we use the generator to process the sequence of values without the need to save the initial values

```python
def f22():
      l=[]
      for i in range(10):
       l.append(i**2)
      return l
```

```python
f22()
```

```python
def f222():
      l=[]
      for i in range(10):
       l.append(i**2)
      yield l
```

```python
f222()
```

```python
next(f222())
```
